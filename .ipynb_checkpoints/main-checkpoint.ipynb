```python
# Import needed libraries
import pickle
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Begin Pipeline

# Read in Images
os.chdir('C:\\Users\\drose\\Google Drive\\mooc\\Udacity\\sdc\\p4\\HiFi-Lane-Tracking')    
import glob
cal_images = []
for img in glob.glob("media\\camera_cal\\*.jpg"):
    n = cv2.imread(img)
    cal_images.append(n)
    
img1  = cal_images[0]
img2 = cal_images[3]

img = img1
"""
Camera Calibration
"""
# Color conversion to Gray
plt.imshow(img)
plt.show()
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.suptitle('gray img 1', fontsize=14, fontweight='bold')
plt.imshow(gray,cmap='gray')
plt.show()

# Prepare object points
objpoints = []
imgpoints = []
nx, ny = 9, 5
objp = np.zeros((nx*ny, 3), np.float32)
objp[:,:2] = np.mgrid[0:nx,0:ny].T.reshape(-1,2) # x,y coordinates

# Find chessboard corners
ret, corners= cv2.findChessboardCorners(gray, (nx,ny), None)

# If found, draw corners
if ret == True:
    # Draw and display the corners
    cv2.drawChessboardCorners(img, (nx, ny), corners, ret)
    imgpoints.append(corners)
    objpoints.append(objp)
    plt.suptitle('cheesboard marked img 1', fontsize=14, fontweight='bold')
    plt.imshow(img)
    plt.show()
    
    ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1], None, None)
    undist = cv2.undistort(img2, mtx, dist, None, mtx)

    plt.suptitle('img 3', fontsize=14, fontweight='bold')
    plt.imshow(img2)
    plt.show()
    plt.suptitle('undistorted img 3', fontsize=14, fontweight='bold')
    plt.imshow(undist)
    plt.show()
    # Choose offset from image corners to plot detected corners
    # This should be chosen to present the result at the proper aspect ratio
    # My choice of 100 pixels is not exact, but close enough for our purpose here
    offset = 200 # offset for dst points
    # Grab the image shape
    img_size = (gray.shape[1], gray.shape[0])

    # For source points I'm grabbing the outer four detected corners
    src = np.float32([corners[0], corners[nx-1], corners[-1], corners[-nx]])
    # For destination points, I'm arbitrarily choosing some points to be
    # a nice fit for displaying our warped result 
    # again, not exact, but close enough for our purposes
    dst = np.float32([[0, 0], # Top-Left
                      [0, 1300], 
                      [700, 1300], 
                      [700, 0]])
    print(np.shape(gray))
    # Given src and dst points, calculate the perspective transform matrix
    M = cv2.getPerspectiveTransform(src, dst)
    # Warp the image using OpenCV warpPerspective()
    warped = cv2.warpPerspective(undist, M, img_size)
    plt.suptitle('transformed img 3', fontsize=14, fontweight='bold')
    plt.imshow(warped)
    plt.show()
if ret == False:
    print('Did not find any corners')


"""
Distortion Correction
"""



"""
Color/Gradient Threshold
Canary, x/y, HSL conversion
"""



"""
Perspective Transform
Birds Eye View
"""



"""
Begin Lane Line Detection
"""



"""
Determine Lane Curvature
"""
```
